```python
import pandas as pd
from math import floor
```

```python
from astropy.coordinates import SkyCoord
import astropy.units as u

# Define the first coordinate
def angler(ra1,dec1,ra2,dec2):
#     ra1 = ra1
#     dec1 = '45d30m00s'
    c1 = SkyCoord(ra1, dec1, frame='icrs')

#     # Define the second coordinate
#     ra2 = '13h00m00s'
#     dec2 = '46d30m00s'
    c2 = SkyCoord(ra2, dec2, frame='icrs')

    # Calculate the angle between the two coordinates
    angle = c1.separation(c2)

    # Print the angle in degrees
    return angle.degree
```

```python
mess=pd.read_csv("coord_mess.csv")
sun=pd.read_csv("Sun_values.csv")
mess=mess.drop(["Unnamed: 0"],axis=1)
# mess=mess.drop([0],axis=0)
mess.iloc[101,0]="M101/102"
mess.iloc[101,0]
```

```python
sun
```

```python
dic={"Messier":[],"Opp_date":[],"Angle":[],"Brightness":[]}
df=pd.DataFrame(dic)
```

```python
for i in range(1,110):
    
    l=list(mess.iloc[i,:])
    ra_h=str(int(floor(float(l[1]))))
    ra_m=str(int(floor(float(l[2]))))
    ra_s=str(int(floor(float(l[3]))))
    dec_d=str(int(floor(float(l[4]))))
    dec_m=str(int(floor(float(l[5]))))
    dec_s=str(int(floor(float(l[6]))))
    mess_ra=ra_h+"h"+ra_m+"m"+ra_s+"s"
    mess_dec=dec_d+"d"+dec_m+"m"+dec_s+"s"
    maxi=0
    ind=-1
    for j in range(364):
        sun_ra=sun.iloc[j,1]
        sun_ra_list=sun_ra.split(':')
        sun_ra_h=str(int(sun_ra_list[0]))
        sun_ra_m=str(int(sun_ra_list[1]))
        sun_ra_s=str(int(sun_ra_list[2]))
        sun_dec=sun.iloc[j,2]
        sun_dec_list=sun_dec.split(':')
        sun_dec_h=str(int(sun_dec_list[0]))
        sun_dec_m=str(int(sun_dec_list[1]))
        sun_dec_s=str(int(sun_dec_list[2]))
        sun_ra=sun_ra_h+"h"+sun_ra_m+"m"+sun_ra_s+"s"
        sun_dec=sun_dec_h+"d"+sun_dec_m+"m"+sun_dec_s+"s"
        ang=angler(mess_ra,mess_dec,sun_ra,sun_dec)
        if float(ang)>maxi:
            maxi=float(ang)
            ind=j
    df.loc[len(df.index)] = [mess.iloc[i,0],sun.iloc[ind,0],maxi,'0']
```
